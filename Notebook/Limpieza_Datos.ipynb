Sabemos que 'Ventas' y 'Beneficio' estan en str hay que cambiardo a float

```python
dataset['Ventas'].dtype
```

Creamos una funcion para quitar el simbolo $ y las comas.

```python
def limpiar_numeros (dato):
    dato = dato.replace('$', '').replace(',', '')
    return float (dato)
```

```python
dataset['Ventas'] = dataset['Ventas'].apply(limpiar_numeros)

dataset['Beneficio'] = dataset['Beneficio'].apply(limpiar_numeros)
```
